```python
df_art = gpd.read_file("../data/EDC_PUBLIC_ART_MONUMENT_WGS84.shp")
```

```python
dest_cols=['poi_id', 'name', 'build_year', 'demolished_year', 'address',
       'latitude', 'longitude', 'external_url', 'image_url', 'heritage_status',
       'current_use', 'poi_type', 'source', 'details', 'dist_start',
       'cleaned_year', 'build_decade', 'poi_type_simple']
```

```python
df_to_db_map={
    'ARTWORKNAM':'name',
    'CREATEYEAR':'build_year'   ,
    'address' :'address' ,
    'details': 'details',
    'image_url':'image_url',
    'heritage_status':'heritage_status',
    'current_use use':'current_use',
    'poi_type':'poi_type',
    'source': 'source', 
    'cleaned_year': 'build_year_clean',
    'build_decade': 'build_decade',
    'poi_type_simple': 'poi_type_simple',
    'LATITUDE': 'latitude',
    'LONGITUDE': 'longitude',
    'ARTIST':'artist'
}
```

```python
df_art.head()
```

```python
def create_detail_desc(row):
    details = "Public Art.  "
    
    if pd.isna(row['MEDIA']) == False:
        details += "\n Media: " + row['MEDIA']
        
    if pd.isna( row['ARTWORKLOC']) == False:   
        details += "\n LOCATION: " + row['ARTWORKLOC'] 
    if pd.isna(row['PARK_NAME']) == False:
        details += "\n PARK NAME: " + row['PARK_NAME']
    if pd.isna(row['NEAR_ST']) == False:
        details += "\n NEAR STREET: " + row['NEAR_ST']
    if pd.isna(row['NEAR_CST']) == False:
        details += "\n NEAR CROSS-STREET: " + row['NEAR_CST']
    return details
```

```python
df_art['POSTAL_CD']=df_art['POSTAL_CD'].fillna(' ')
```

```python

df_art['address'] = df_art['ADD_FULL'] + " Toronto ON " + df_art['POSTAL_CD']
df_art['source']= "https://www.toronto.ca/city-government/data-research-maps/open-data/"
df_art['poi_type_simple']='Art'
df_art['poi_type']="Public Art, " + df_art['MEDIA'] 
df_art['details']=df_art.apply(lambda row: create_detail_desc(row), axis=1)
```

```python
df_art.head()
```

```python
new_df=[]
for index, row in df_art.iterrows():
    poi_dict ={df_to_db_map[k]:v for k, v in row.items() if k in df_to_db_map.keys() and not pd.isnull(v)}
    new_df.append(poi_dict)
```

```python
df_art2 = pd.DataFrame(new_df)
```
